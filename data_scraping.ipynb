```python
#importing the libraries
from bs4 import BeautifulSoup
import requests
```

```python
#dataset link
url = 'https://books.toscrape.com/'
#store information in a variable from the link
response = requests.get(url)
#handle errors and exceptions
response.encoding = 'utf-8'
#extract the info in the web page as text
soup = BeautifulSoup(response.text,  'lxml')
```

```python
#to make the results or the response visually appealing
print(soup.prettify)
```

```python
#printing all books
books= soup.find_all('article', class_ = 'product_pod')
#finding the title of the books
for book in books:
 book_title = book.h3.a.get('title')
 #accessing the rating using the variable,class and the position of the 5 in  the rating list
 rating = book.find('p', class_ = 'star-rating').get('class')[1]
 #getting the prices of the books and put in text format
 price = book.find('p', class_ = 'price_color').text
 #getting book_link_URL
 book_link_URL = url + book.h3.a.get('href')
 #Filtering to only obtain books that will be appealing to the book lovers by creating a condition
 if rating == 'Five' or rating == 'Four':
   print(f'''
   Book Recommendation:
   Title: "{book_title}"
   Rating: {rating} star
   Price: {price}
   Link: {book_link_URL}

   I want to purchase this book because it has a high rating.

  ''')
 else:
    print(f'''
    Book Review:
    Title: "{book_title}"
    Rating: {rating} star
    Price: {price}
    Link: {book_link_URL}

    I do not recommend this book because the rating is below 4 stars.
    ''')
```
